```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Turn off pandas warnings
warnings.filterwarnings('ignore', category=Warning)
```

```python
data = pd.read_excel('immigration.xlsx')
data.index = data['TIME']
data.drop(columns=['TIME'], inplace=True)
```

```python
data1 = data.copy()
data1['sum'] = data1.sum(axis=1)
data1 = data1.sort_values(by=['sum'], ascending=False)
```

```python
data2 = data1.loc[data1.index.isin(['European Union', 'Germany', 'Spain', 'France', 'Poland'])]
data2.drop(columns=['sum'], inplace=True)
data2
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
x = data2.columns

sns.lineplot(x=x, y=data2.loc['European Union'].values, color='#2ECC71')

plt.title('European Union')
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
```

```python
countries = ['Germany', 'Spain', 'France', 'Poland']
colours = ['#3498DB', '#F1C40F', '#8E44AD', '#E74C3C']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle('Immigration Over Time for Different Countries', fontsize=16)

# Loop through countries and plot each one on a separate subplot
for i, (name, color) in enumerate(zip(countries, colours)):
    row = i // 2
    col = i % 2
    ax = axes[row, col]

    sns.lineplot(x=x, y=data2.loc[name].values, color=color, ax=ax)

    ax.set_title(name)
    ax.set_xlabel('Year')
    ax.set_ylabel('Number of immigrants')

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```
